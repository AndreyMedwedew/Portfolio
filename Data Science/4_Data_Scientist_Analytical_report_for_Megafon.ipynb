`megafon.csv` содержит следующие значения: <br><br>
&nbsp;&nbsp;&nbsp;&nbsp; `user_id` — идентификатор абонента;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Q1` — ответ на первый вопрос;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Q2` — ответ на второй вопрос;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Total Traffic(MB)` — объем трафика передачи данных <sup>1 </sup>; <br>
&nbsp;&nbsp;&nbsp;&nbsp; `Downlink Throughput(Kbps)` — средняя скорость «к абоненту» <sup>2 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Uplink Throughput(Kbps)`— средняя скорость «от абонента» <sup>3 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Downlink TCP Retransmission Rate(%)` — частота переотправок пакетов «к абоненту» <sup>4 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Video Streaming Download Throughput(Kbps)` — скорость загрузки потокового видео <sup>5 </sup>;<br>

<sup>1 </sup> — Насколько активно абонент использует мобильный интернет.<br>
<sup>2 </sup> — Считается по всему трафику передачи данных.<br>
<sup>3 </sup> — Считается по всему трафику передачи данных.<br>
<sup>4 </sup> — Чем выше, тем хуже. Если в канале возникает ошибка, пакет переотправляется. Снижается полезная скорость.<br>
<sup>5 </sup> — Чем выше, тем лучше — меньше прерываний и лучше качество картинки.<br>
<sup>6 </sup> — Сколько времени пройдёт между нажатием на кнопку Play и началом воспроизведения видео. Чем меньше это время, тем быстрее начинается воспроизведение.<br>
<sup>7 </sup> — Чем выше, тем лучше.<br>
<sup>8 </sup> — Чем меньше, тем лучше — быстрее загружаются web-страницы.<br>

Первый технический показатель представлен как сумма за период в одну неделю перед участием в опросе. Остальные технические показатели отображают среднее значение по данному признаку за период в одну неделю перед участием в опросе.

# 3 Блок со статистическим тестом

Выберите метод для статистической проверки вашего предположения, письменно обоснуйте свой выбор. Проверьте, удовлетворяет ли выборка допущениям (assumptions) выбранного вами теста. Реализуйте тест и на основании результатов сформируйте вывод.

```python
# Напишем функцию, которая будет увеличивать нашу выборку с помощью бутстрепа

def bootstrap(arr, n=100, N_TRIAL = 10000, func = np.mean):
  res = []
  n = min(len(arr), n)
  # Инициализируем цикл попыток
  for _ in range(N_TRIAL):
    # Делаем подвыборку размера n
    if len(arr.shape) > 1:
      ind = np.random.choice(arr, size = (n,))
      subsample = arr.iloc[ind]
    else:
      subsample = np.random.choice(arr, size = (n,))
    # рассчитываем статистику
    stat = func(subsample)
    res.append(stat)
  return res # shape (N_TRIAL,)
```

Теперь посмотрим на расположения 95%ых доверительных интервалов для разного типа клиентов

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
fig, axs = plt.subplots(2, 4, figsize=(20, 9))


a = 0
for i in df_dissatisfied_clients.columns[3:7]:

  samples_a = bootstrap(df_satisfied_clients[i], \
                      n = df_satisfied_clients[i].size)
  samples_b = bootstrap(df_dissatisfied_clients[i], \
                      n = df_dissatisfied_clients[i].size)

  ax = sns.distplot(samples_a, ax = axs[0][a], label = 'Довольные')
  ax = sns.distplot(samples_b, ax = axs[0][a], label = 'Недовольные')
  ax.legend()
  ax.set_title(i)
  ax.axvline(np.percentile(samples_a,2.5),linestyle = '-',color='r')
  ax.axvline(np.percentile(samples_a,97.5),linestyle = '-',color='g')

  ax.axvline(np.percentile(samples_b,2.5),linestyle = '--',color='r')
  ax.axvline(np.percentile(samples_b,97.5),linestyle = '--',color='g')

  a += 1




a = 0
for i in df_dissatisfied_clients.columns[7:]:

  samples_a = bootstrap(df_satisfied_clients[i], \
                      n = df_satisfied_clients[i].size)
  samples_b = bootstrap(df_dissatisfied_clients[i], \
                      n = df_dissatisfied_clients[i].size)

  # fig, ax = plt.subplots(figsize=(6, 3))
  ax = sns.distplot(samples_a, ax = axs[1][a], label = 'Довольные')
  ax = sns.distplot(samples_b, ax = axs[1][a], label = 'Недовольные')
  ax.legend()
  ax.set_title(i)
  ax.axvline(np.percentile(samples_a,2.5),linestyle = '-',color='r')
  ax.axvline(np.percentile(samples_a,97.5),linestyle = '-',color='g')

  ax.axvline(np.percentile(samples_b,2.5),linestyle = '--',color='r')
  ax.axvline(np.percentile(samples_b,97.5),linestyle = '--',color='g')

  a += 1
```

```python
# Группировка причин недовольных клиентов
list_df_dissatisfied_clients = []
for i in df_dissatisfied_clients['Q2']:
  if type(i) == str and len(i) > 1:
    for j in i.split(', '):
      list_df_dissatisfied_clients.append(int(j))
  else:
    list_df_dissatisfied_clients.append(int(i))
```

```python
# Считаем количество оценок
counts = {}
for i in list_df_dissatisfied_clients:
  counts[i] = counts.get(i, 0) + 1
```

```python
# Так как у нас оценок 10 и 34 при втором вопросе не предполагалось, думаю, что клиент имел ввиду две оценки вместо одной 1, 0 и 3, 4 соответственно
# Удалим некорректные оценки и добавим их в виде обычных оценок.

counts = counts[:-2]
counts.loc[0] = counts.loc[0] + 1
counts.loc[1] = counts.loc[1] + 1
counts.loc[3] = counts.loc[3] + 1
counts.loc[4] = counts.loc[4] + 1
counts.set_index(pd.Index(['Плохое качество связи в зданиях,  ТЦ и т.п.',
                           'Недозвоны, обрывы при звонках',
                           'Не стали отвечать на 2ой вопрос',
                           'Медленный мобильный интернет',
                           'Медленная загрузка видео',
                           'Время ожидания гудков при звонке',
                           'Свой вариант',
                           'Затрудняюсь ответить']), inplace=True)
```
