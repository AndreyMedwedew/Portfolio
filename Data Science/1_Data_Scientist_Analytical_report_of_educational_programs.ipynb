```python
main = main_original[main_original['is_homework']==True].copy()
main['start_date'] = pd.to_datetime(main.start_date)
main['finish_date'] = pd.to_datetime(main.finish_date)

main['YearMonth']= main['start_date'].values.astype('datetime64[M]')
main['YearMonth'] = main['YearMonth'].dt.to_period('M')

# первая дз на курсе
n_of_1st_hw = main.groupby(['title']).agg({'module_number':['min']}).reset_index()
n_of_1st_hw.columns = n_of_1st_hw.columns.droplevel(0)
n_of_1st_hw.columns = ['title', 'module_number']

# сортируем дф с помощью объединения с дф только с номерами первых дз на каждом курсе
main_sorted = main.merge(n_of_1st_hw, left_on=['title','module_number'], right_on=['title','module_number'])
main_sorted = main_sorted.groupby(['title', 'YearMonth'])['YearMonth'].count().unstack(fill_value=0).stack().to_frame().reset_index()

main_sorted.columns = ['title','year_and_month','growth']
main_sorted.head(15)
```

4. **(p)**
Рассчитать количество прогрессов по выполнению домашних работ в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно) для каждого курса. Учитывать, что выполнение домашнего задания может перетекать из одного месяца в другой (такие дз надо включать в общее число прогрессов для всех месяцев, которые покрывает срок выполнения этих дз)

```python

main = main_original[main_original['is_homework']==True][['title','progress_id', 'lesson_token', 'start_date','finish_date']].copy()
main['start_date'] = pd.to_datetime(main.start_date, utc = True)
main['finish_date'] = pd.to_datetime(main.finish_date, utc = True)

# делаем так, чтобы в start_date стоял первый день месяца, в котором начата дз
# а в finish_date последний день месяца начала, если работа еще не закончена, и то, что есть, если закончена
from pandas.tseries.offsets import MonthEnd, MonthBegin

# приводим все значения к первой дате месяца
main['start_date'] = main['start_date'].apply(lambda x: pd.Timestamp(x) + Day() - MonthBegin())


main['finish_date'] = main[['start_date','finish_date']].apply(lambda x: pd.Timestamp(x['start_date']) - Day() + MonthEnd() if pd.isna(x['finish_date']) else x['finish_date'],axis=1)


main['start_date'] = pd.to_datetime(main.start_date, utc = True)
main['finish_date'] = pd.to_datetime(main.finish_date, utc = True)

# делаем список месяцев, в которых выполнялось каждое дз
main['time']=main[['start_date','finish_date']].apply(lambda x: list(pd.date_range(start=x['start_date'], periods=ceil((x['finish_date'] - x['start_date'])/np.timedelta64(1, 'M')), freq='M')),axis=1)

final_df = main.explode('time')

final_df['time']= final_df['time'].dt.to_period('M')

workload = final_df.groupby(['title','time'])['title'].count().unstack(fill_value=0).stack().to_frame().reset_index()

workload.columns = ['title','year_and_month', 'workload']
workload.head(30)
```

## 3. Выявление проблемных модулей

1. **(p)** Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля (разность между временем начала и окончания выполнения домашней работы) для каждого курса. Если домашних заданий в модуле несколько, то считать разность между временем начала выполнения первой домашней работы и временем окончания выполнения последней домашней работы в модуле

```python
# создаем дф с указанием количества частей дз в каждом модуле – hw_num
course_all_hw = courses_all[courses_all['is_homework']==True].copy()
hw_num = course_all_hw.pivot_table(index=['title','module_number'],values='lesson_number',aggfunc={'lesson_number':'nunique'})
hw_num = hw_num.reset_index().rename(columns={'lesson_number':'hw_num'})
```

```python
# создаем дф, где для каждой работы выставлено количество начатых частей дз, мин время начала дз, макс вермя окончания дз
main = main_original[(main_original['is_homework']==True)&(main_original['status']=='done')].copy()

hw_agg_values = main.groupby(['title','module_title','module_number','progress_id']).agg({'lesson_number':'count','start_date':'min','finish_date':'max'}).reset_index()

# убираем некорректные значения
hw_agg_values = hw_agg_values[(hw_agg_values['start_date']<hw_agg_values['finish_date'])]
main = hw_num.merge(hw_agg_values) # объединяем с первым дф

# оставляем только те дз, где количество начатых частей дз совпадает с общим количеством частей дз
main = main[main['lesson_number']==main['hw_num']]
```

## 4. Расчет конверсии

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

```python
main = main_original[main_original['is_homework']==True].copy()

# если есть запись по дз – значит, к этому дз приступили
started_hws = main.groupby(['title','module_title','module_number']).agg({'student_id':'nunique'}).reset_index()

# законченные работы
done_hws = main[main['status']=='done'].groupby(['title','module_title','module_number']).agg({'student_id':'nunique'}).reset_index()


conversion = started_hws.merge(done_hws, left_on=['title','module_title','module_number'], \
               right_on=['title','module_title','module_number'], how='left') \
                .rename(columns={'student_id_x':'started_hws','student_id_y':'done_hws'}) \
                .dropna() \
                .astype({'done_hws': 'int64'}) \
                .sort_values(by=['title','module_number','module_title'])

```

```python
conversion = conversion.groupby(['title']).apply(lambda x: x['started_hws']/x['done_hws'].shift(periods=1)).to_frame().reset_index().drop('level_1',axis=1)
conversion.columns = ['title','conversion_rate']

conversion
```
